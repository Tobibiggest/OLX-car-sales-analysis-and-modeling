# Feature Engineering:Creating features


 * Using Air Condition as the most important car feature, let's see number of cars with the feature Air condition.
 * We then create a new colunm for car feature count - tells the number of features each car has.

```python
# Define a function to count a data point in a row of a column

def count_data_point(row, column, data_point):
    
    # Split the column value by comma and strip whitespace
    
    values = row[column].split(",")
    values = [value.strip() for value in values]
    
    # Return the count of the data point in the values list
    
    return values.count(data_point)

# Apply the function to each row of the dataframe



data["AC_count"] = data.apply(count_data_point, axis=1, column="Car_Features", data_point="Air Conditioning")

# Print the dataframe

print(data)
```

```python
# create a new colunm for car feature count - tells the number of features each car has.

data['Car_Features_count'] = data['Car_Features'].str.split(',').apply(len)

# Dropping the car feature colunm

data = data.drop(['Car_Features'], axis = 1)
data.head()
```

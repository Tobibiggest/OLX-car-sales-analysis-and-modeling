```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import sklearn
```

```python
data = pd.read_csv("C:/Users/Biggest/Downloads/OLX_cars_dataset00.csv")
```

```python
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
# create a new colunm for car feature count - tells the number of features each car has.

data['Car_Features_count'] = data['Car Features'].str.split(',').apply(len)


data.head()
```

```python
x = data[["KM's driven", 'Car_Age', 'Car_Features_count']]
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stndrd = StandardScaler()
x_train = stndrd.fit_transform(x_train)
x_test = stndrd.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```
